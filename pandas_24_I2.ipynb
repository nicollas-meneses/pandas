```python
# Filtrando dados do dataframe através de Series pandas
# Mostre todos as moradias cujo valor total seja inferior a 4000
# e sejam do tipo Apartamento:
c1 = df["total"] < 4000
c2 = df["tipo"] == "Apartamento"
```

```python
df[c1 & c2]
```

```python
# Mostre todas as moradias que são do bairro Mooca
# Ou possuem valor total < 7000
c1 = df["bairro"] == 'Mooca'
c2 = df["total"] < 3000
```

```python
df[c1 | c2]
```

```python
# Realizando filtros através do método query()
# Mostre todas as casas em condomínio que possuem valor total
# até 5000
df.query("tipo == 'Casa em condomínio' and total <= 5000")
```

```python
# Mostre o valor médio pago em casas de condomínio cujo valor total
# seja até 5000
df.query("tipo == 'Casa em condomínio' and total <= 5000")["total"].mean()
```

```python
# Mediana:
df.query("tipo == 'Casa em condomínio' and total <= 5000")["total"].median()
# Desvio padrão:
df.query("tipo == 'Casa em condomínio' and total <= 5000")["total"].std()
```

```python
# Mostre todas as moradias que não sejam Apartamentos
# e cujo valor total está entre 1000 e 2000
df.query("total >= 1000 and total <= 2000 and tipo != 'Apartamento'")
```

```python
# Análise estatística geral:
df.describe()
```

```python
# Normalizando dados nulos:
df.fillna({
    "bairro":"NA",
    "total":0,
    "quartos":1,
    "endereco":"NA"
}, inplace=True)
```

```python
# Exibindo valores nulos:
# ~ é o operador not usado nas Series do Pandas
df[~df["total"].notnull()].count()
```

```python
# Adicionando coluna ao dataframe:
# Adicionar coluna "condominio" que será a diferença entre
# o valtor total e o aluguel:
df["condominio"] = df["total"] - df["aluguel"]
```

```python
df
```

```python
# Criando coluna condicional
lista_desconto = []
```

```python
# Através do comando 'for' iremos percorrer cada linha da coluna
# 'total':
for linha in df["total"]:
  if linha >= 15000:
    lista_desconto.append(linha * 0.95)
  else:
    lista_desconto.append(linha)
```

```python
# Incluindo nova coluna condicional através da lista
df["desconto"] = lista_desconto
```

```python
df.info()
```

```python
# Uso de função lambda com map() e if ternário:
# valor if True else valor
df["teste"] = list(map(lambda a: a * 0.95 if a >= 15000 else a, df["total"]))
```

```python
# Deletar coluna:
# axis=1 indica que queremos deletar a coluna inteira:
df.drop("teste", axis=1, inplace=True)
```

```python
# Criando gráficos no pandas
# Usamos o método plot(). Por padrão (atualmente) a função plot()
# usa a estrutura do matplotlib para construtir os gráficos:
df.plot()
```

```python
grafico1 = df.groupby("tipo")["total"].median().sort_values(ascending=False)
```

```python
# Grafíco de barras horizontal
grafico1.plot(
    kind="barh",
    figsize=(5, 3),
    title="Mediana de valores por tipo de moradia",
    xlabel="Mediana",
    ylabel="Tipo de moradia",
    color="darkblue"
)
```

```python
# Grafíco de barras padrão
grafico1.plot(
    kind="bar",
    figsize=(5, 3),
    title="Mediana de valores por tipo de moradia",
    xlabel="Tipo de moradia",
    ylabel="Mediana",
    color="green"
)
```

```python

grafico2 = df.groupby("tipo")["tipo"].count().sort_values(ascending=False)
```

```python
# Gráfico de pizza com valores percentuais:

grafico2.plot(
    kind="pie",
    colormap="Pastel1",
    figsize=(5, 3),
    title="Percentual de moradias por tipo",
    autopct="%1.2f%%",
    # fontsize=7
)
```
